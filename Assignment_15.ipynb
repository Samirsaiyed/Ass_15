7.Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to
its next() method: 2, 3, 5, 7, 11, ...

```python
def genPrimes():
    n = 0
    while True :
        if n == 2 or n == 3 :
            yield n
        elif ((n-1)%6 == 0 or (n+1)%6 == 0) and n !=1:
            yield n
        n = n+1
output = genPrimes()
for ele in range(5):
    print(next(output))
```
